# Seaborn Overview

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.




## Distribution Plots

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

## Imports

```python
import seaborn as sns  # import seaborn for data visualization
%matplotlib inline    # show plots directly inside the notebook
```

## Data
Seaborn comes with built-in data sets!

```python
tips = sns.load_dataset('tips')  # load the built-in tips dataset from seaborn
```

```python
tips.head()  # show the first 5 rows of the dataset
```

## distplot

The distplot shows the distribution of a univariate set of observations.

```python
sns.distplot(tips['total_bill'])  # plot the distribution of total bill amounts
# Safe to ignore warnings
```

To remove the kde layer and just have the histogram use:

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)  # show just the histogram with 30 bars, no smooth curve
```

## jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')  # plot total_bill vs tip as a scatter plot with distributions on the sides
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')  # same plot but use hexagonal bins to show where most points cluster
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')  # same plot but add a regression line to see the trend
```

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(tips)  # plot every numeric column against every other numeric column
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')  # same pairplot but color the points by gender
```

## rugplot

rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:

```python
sns.rugplot(tips['total_bill'])  # draw a small tick mark for each data point along the axis
```

## kdeplot

kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np  # import numpy for numerical operations
import matplotlib.pyplot as plt  # import matplotlib for drawing plots
from scipy import stats  # import stats for creating probability distributions

#Create dataset
dataset = np.random.randn(25)  # create 25 random numbers following a normal distribution

# Create another rugplot
sns.rugplot(dataset);  # draw a small tick for each data point

# Set up the x-axis for the plot
x_min = dataset.min() - 2  # lowest x value with a little padding on the left
x_max = dataset.max() + 2  # highest x value with a little padding on the right

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)  # create 100 evenly spaced points across the x range

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'  # link for more info on bandwidth

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2  # calculate best bandwidth using Silverman's rule


# Create an empty kernel list
kernel_list = []  # empty list to store each individual kernel curve

# Plot each basis function
for data_point in dataset:  # loop through every data point one by one
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)  # create a bell curve centered on this data point
    kernel_list.append(kernel)  # save this kernel to the list
    
    #Scale for plotting
    kernel = kernel / kernel.max()  # normalize so the tallest peak equals 1
    kernel = kernel * .4  # shrink it down so it fits nicely on the chart
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)  # draw each kernel in light grey

plt.ylim(0,1)  # set the y-axis to go from 0 to 1
```

```python
# To get the kde plot we can sum these basis functions.

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)  # add all the individual kernels together into one curve

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')  # draw the final KDE curve in red

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')  # draw the original data ticks in the same red color

# Get rid of y-tick marks
plt.yticks([])  # remove the numbers from the y axis to keep it clean

# Set title
plt.suptitle("Sum of the Basis Functions")  # add a title to the figure
```

So with our tips dataset:

```python
sns.kdeplot(tips['total_bill'])  # draw a smooth density curve for total bill amounts
sns.rugplot(tips['total_bill'])  # show individual data points as ticks below the curve
```

```python
sns.kdeplot(tips['tip'])  # draw a smooth density curve for tip amounts
sns.rugplot(tips['tip'])  # show each tip value as a tick below the curve
```

# Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Let's go through examples of each!

```python
import seaborn as sns  # import seaborn for visualization
%matplotlib inline  # display plots directly in the notebook
```

```python
tips = sns.load_dataset('tips')  # load the tips dataset again
tips.head()  # show the first 5 rows to preview the data
```

## barplot and countplot

These very similar plots allow you to get aggregate data off a categorical feature in your data. **barplot** is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
sns.barplot(x='sex',y='total_bill',data=tips)  # compare the average total bill between males and females
```

```python
import numpy as np  # import numpy for math functions like standard deviation
```

You can change the estimator object to your own function, that converts a vector to a scalar:

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)  # show how spread out the bill amounts are for each gender instead of the average
```

### countplot

This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value:

```python
sns.countplot(x='sex',data=tips)  # count how many male and female customers there are and show as bars
```

## boxplot and violinplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')  # show the spread of total bills for each day of the week
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')  # show horizontal box plots for all numeric columns in the dataset
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")  # split each day's box plot by whether the customer smokes or not
```

### violinplot
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')  # show the shape of the bill distribution for each day using a violin plot
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')  # split each violin into male and female to compare genders per day
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')  # merge male and female into one split violin so they share the same shape
```

## stripplot and swarmplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.stripplot(x="day", y="total_bill", data=tips, palette='rainbow')  # show every individual bill as a dot, grouped by day
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True, palette='rainbow')  # add random side-to-side shift so overlapping dots are easier to see
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')  # color each dot by gender so you can see male vs female patterns
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',dodge=True)  # separate male and female dots into their own columns side by side
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)  # spread dots apart so none overlap, giving a clearer picture of the full distribution
```

### Combining Categorical Plots

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')  # draw a violin to show tip distribution for each day
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)  # overlay black dots on the violin to show actual data points
```

## catplot

factorplot is the most general form of a categorical plot. It can take in a kind parameter to adjust the plot type:

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar')  # create a bar chart comparing total bill by gender using the general categorical plot function
```

# Matrix Plots

Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data (later in the machine learning section we will learn how to formally cluster data).

Let's begin by exploring seaborn's heatmap and clutermap:

```python
import seaborn as sns  # import seaborn for visualization
%matplotlib inline  # display plots inside the notebook
```

## Heatmap

In order for a heatmap to work properly, your data should already be in a matrix form, the sns.heatmap function basically just colors it in for you. For example:

```python
tips.head()  # show the tips dataframe again as a reminder before using it
```

```python
# Matrix form for correlation data
tips.corr()  # calculate how strongly each numeric column relates to every other column
```

```python
sns.heatmap(tips.corr())  # visualize the correlation table as a color-coded grid
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True)  # same heatmap but with a blue-red color scale and the numbers shown inside each cell
```

Or for the flights data:

```python
flights.pivot_table(values='passengers',index='month',columns='year')  # reshape the flights data so rows are months and columns are years, values are passenger counts
```

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')  # save the reshaped table to a variable
sns.heatmap(pvflights)  # draw a heatmap showing passenger counts by month and year
```

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1)  # same heatmap with a dark magma color scheme and white lines separating the cells
```

## clustermap

The clustermap uses hierarchal clustering to produce a clustered version of the heatmap. For example:

```python
sns.clustermap(pvflights)  # group similar months and years together and draw a rearranged heatmap based on similarity
```

Notice now how the years and months are no longer in order, instead they are grouped by similarity in value (passenger count). That means we can begin to infer things from this plot, such as August and July being similar (makes sense, since they are both summer travel months)

```python
# More options to get the information a little clearer like normalization
sns.clustermap(pvflights,standard_scale=1)  # same clustermap but normalize each column so values range from 0 to 1, making patterns clearer
```

# Regression Plots

Seaborn has many built-in capabilities for regression plots, however we won't really discuss regression until the machine learning section of the course, so we will only cover the **lmplot()** function for now.

**lmplot** allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

Let's explore how this works:

```python
import seaborn as sns  # import seaborn for visualization
%matplotlib inline  # show plots inside the notebook
```

```python
tips = sns.load_dataset('tips')  # load the tips dataset
```

```python
tips.head()  # show the first 5 rows of the data
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips)  # plot total_bill vs tip and draw a linear regression line to show the trend
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')  # same regression plot but color the dots differently for male and female
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')  # same plot but apply a blue-red color palette to differentiate genders
```

## Using a Grid

We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')  # create a separate regression plot for males and females side by side
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)  # create a grid of plots split by gender (rows) and meal time (columns)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')  # one regression plot per day, colored by gender
```

## Aspect and Size

Seaborn figures can have their size and aspect ratio adjusted with the **height** and **aspect** parameters:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,height=8)  # same daily plots by gender but taller and narrower to fit better
```
