## 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

## 2. Load Dataset

```python
df = pd.read_csv("heart_disease_uci.csv")
df.head()
```

## 3. Check Missing Values

```python
print(df.isna())
print(df.isna().sum())
```

## 4. Check Duplicate Records

```python
df.duplicated()[df.duplicated()==True]
```

## 5. Dataset Shape

```python
df.shape
```

## 6. Dataset Information

```python
df.info()
```

## 7. Statistical Summary

```python
df.describe()
```

## 8. Target Variable Distribution

```python
sns.countplot(x='num', data=df)
plt.title("Heart Disease Distribution")
plt.show()
```

## 9. Gender vs Heart Disease

```python
sns.countplot(x='sex', hue='num', data=df)
plt.title("Gender vs Heart Disease")
plt.show()
```

## 10. Chest Pain vs Heart Disease

```python
sns.countplot(x='cp', hue='num', data=df)
plt.title("Chest Pain vs Heart Disease")
plt.show()
```

## 11. Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

## 12. Correlation Heatmap

```python
corr = df.select_dtypes(include="number").corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## 13. Observations

- Majority of patients are between 4060 years old  
- Male patients show higher heart disease cases  
- Chest pain type strongly affects heart disease  
- Age shows moderate correlation with target
